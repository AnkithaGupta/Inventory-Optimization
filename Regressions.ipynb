```python
df_logistic['others_class'] = df_logistic['Vest']+ df_logistic['Bodysuits']+ df_logistic['Kimono']+ df_logistic['traditional_cardigan']
df_logistic.drop(['Vest', 'Bodysuits', 'Kimono', 'traditional_cardigan'], axis=1, inplace=True)

df_logistic['cardigans'] = df_logistic['long_cardigans']+ df_logistic['short_cardigans']
df_logistic.drop(['long_cardigans', 'short_cardigans'], axis=1, inplace=True)

df_logistic['sweater'] = df_logistic['long_sleeve_knits'] + df_logistic['pullover_sweater']+ df_logistic['Sweatshirt']
df_logistic.drop(['long_sleeve_knits', 'pullover_sweater', 'Sweatshirt'], axis=1, inplace=True)

df_logistic['tanks'] = df_logistic['sleeveless_woven_tanks'] + df_logistic['sweater_tanks_and_tanks'] + df_logistic['short_sleeve_knits']+ df_logistic['misc_fashion_tops']
df_logistic.drop(['sleeveless_woven_tanks', 'sweater_tanks_and_tanks', 'short_sleeve_knits','misc_fashion_tops'], axis=1, inplace=True)
```

```python
def group_colors(color):
    neutral = ['Ivory', 'White', 'Sand', 'Heather Oat', 'Heather Gray', 'Gray', 'Black', 'Black/White', 'Cinnamon', 'Taupe', 'Natural', 'Tan', 'Charcoal', 'Silver', 'No Color']
    earthy = ['Brick', 'Rust', 'Marigold', 'Moss', 'Olive', 'Dark Olive', 'Amber', 'Mustard', 'Cognac']
    blue = ['Navy', 'Baby Blue', 'Chambray', 'Indigo', 'Cobalt Blue', 'Teal', 'Aqua', 'Turquoise', 'Light Blue']
    green = ['FOREST GREEN', 'Hunter Green', 'Evergreen', 'Pine']
    pink_red = ['Rose', 'Bright Red', 'Coral', 'Fuchsia', 'Raspberry', 'Living Coral', 'Red', 'Deep Moss', 'Dark Rose']
    purple = ['Purple', 'Vintage Purple']
    yellow_orange = ['Sunshine', 'Lemon', 'Orange', 'Tangerine', 'Banana', 'Saffron', 'Adobe', 'Peach', 'Gold', 'Camel', 'Champagne', 'Yellow']
    other = ['Lavender', 'Lite', 'Ink Navy', 'Multi', 'Blush', 'Med', 'Burgundy', 'Neon Coral', 'Brown', 'Emerald', 'Magenta', 'Mauve', 'Dark', 'Topaz', 'Mint', 'Sage', 'Berry']
    
    if color in neutral:
        return 'Neutral'
    elif color in earthy:
        return 'Earthy'
    elif color in blue:
        return 'Blue'
    elif color in green:
        return 'Green'
    elif color in pink_red:
        return 'Pink/Red'
    elif color in purple:
        return 'Purple'
    elif color in yellow_orange:
        return 'Yellow/Orange'
    else:
        return 'Other_Color'
   
```

```python
df_logistic['color_group'] = df_logistic['color'].apply(group_colors)
```

```python
# Create dummies for the 'color' column
color_dummies = pd.get_dummies(df_logistic['color_group'])
df_logistic = pd.concat([df_logistic, color_dummies], axis=1)
df_logistic.drop('color_group', axis=1, inplace=True)
df_logistic.drop('color', axis=1, inplace=True)
```

```python
#Using Size_Nosize, Fall, Tanks and Other colors as base categories
x=df_logistic[['gross_margin','size_XXS','size_L','size_M','size_S', 'size_XS','size_XL','size_S/M','size_M/L', 'Spring','Summer','Winter','Blouse','button_down', 'others_class', 'cardigans', 'sweater', 'Blue', 'Earthy', 'Green', 'Other_Color', 'Pink/Red', 'Purple', 'Yellow/Orange']]
y=df_logistic['product clearance']

##train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state=101)

#Implementing the Log Model
from sklearn.linear_model import LogisticRegression #import
logmodel=LogisticRegression(solver='liblinear') #initialize
logmodel.fit(x_train,y_train)#fit

logmodel.intercept_ #bo
logmodel.coef_ #b1

probabilities=logmodel.predict_proba(x_test)
y_pred=logmodel.predict(x_test)


##evaluate the model using f1-score
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
from scipy.stats import norm
import numpy as np
# create a DataFrame to store the feature names, coefficients, and p-values
coef_df = pd.DataFrame({'feature': x.columns, 'coefficient': logmodel.coef_[0]})

# calculate the p-values for each coefficient using the normal distribution
coef_df['p-value'] = [2 * (1 - norm.cdf(np.abs(x))) for x in logmodel.coef_[0]]

# print the coefficients, p-values, and significance levels (alpha=0.05) next to the feature names
for idx, row in coef_df.iterrows():
    if row['p-value'] < 0.05:
        signif = '*'
    else:
        signif = ''
    print(f"{row['feature']}: {row['coefficient']:.3f} (p-value={row['p-value']:.3f}{signif})")
```

```python
#Using the confusion matrix
from sklearn.metrics import (accuracy_score, precision_score,recall_score,f1_score
                             ,confusion_matrix)
confusion_mat = pd.DataFrame(confusion_matrix(y_test,y_pred), 
                             index = ['Actual:0', 'Actual:1'],
                             columns = ['Pred:0', 'Pred:1'])

confusion_mat
```

```python
from scipy.stats import norm
import numpy as np
# create a DataFrame to store the feature names, coefficients, and p-values
coef_df = pd.DataFrame({'feature': x.columns, 'coefficient': logmodel.coef_[0]})

# calculate the p-values for each coefficient using the normal distribution
coef_df['p-value'] = [2 * (1 - norm.cdf(np.abs(x))) for x in logmodel.coef_[0]]

# print the coefficients, p-values, and significance levels (alpha=0.05) next to the feature names
for idx, row in coef_df.iterrows():
    if row['p-value'] < 0.05:
        signif = '*'
    else:
        signif = ''
    print(f"{row['feature']}: {row['coefficient']:.3f} (p-value={row['p-value']:.3f}{signif})")
```
